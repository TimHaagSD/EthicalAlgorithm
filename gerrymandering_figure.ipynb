```python
# Imports
import pandas as pd
#import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import random
import math
```

```python
# Plot values and parameters
np.random.seed(5)
x, y = np.random.random(20), np.random.random(20)
a, b = np.random.random(20), np.random.random(20)
c, d = np.random.random(20), np.random.random(20)
e, f = np.random.random(20), np.random.random(20)

admitted = 5
c1 = '#8da0cb'  # Blue
c2 = '#fc8d62'  # Orange
```

```python
# Create figure
fig, axs = plt.subplots(2, 2, sharey=True, sharex=True, figsize=(8,8))
axs[0, 0].scatter(x[admitted+1:], y[admitted+1:], c=c1, marker='o')
axs[0, 0].scatter(x[:admitted+1], y[:admitted+1], c=c2, marker='o')
axs[0, 0].set_title('Male/Circle')
axs[0, 1].scatter(a[admitted+1:], b[admitted+1:], c=c1, marker='o')
axs[0, 1].scatter(a[:admitted+1], b[:admitted+1], c=c2, marker='o')
axs[0, 1].set_title('Female/Circle')
axs[1, 0].scatter(c[admitted+1:], d[admitted+1:], c=c1, marker='s')
axs[1, 0].scatter(c[:admitted+1], d[:admitted+1], c=c2, marker='s')
axs[1, 0].set_title('Male/Square')
axs[1, 1].scatter(e[admitted+1:], f[admitted+1:], c=c1, marker='s')
axs[1, 1].scatter(e[:admitted+1], f[:admitted+1], c=c2, marker='s')
axs[1, 1].set_title('Female/Square')
```

```python
# Create figure
fig, axs = plt.subplots(2, 2, sharey=True, sharex=True, figsize=(8,8))
axs[0, 0].scatter(x[admitted*2+1:], y[admitted*2+1:], c=c1, marker='o')
axs[0, 0].scatter(x[:admitted*2+1], y[:admitted*2+1], c=c2, marker='o')
axs[0, 0].set_title('Male/Circle')
axs[0, 1].scatter(a, b, c=c1, marker='o')
axs[0, 1].set_title('Female/Circle')
axs[1, 0].scatter(c, d, c=c1, marker='s')
axs[1, 0].set_title('Male/Square')
axs[1, 1].scatter(e[admitted*2+1:], f[admitted*2+1:], c=c1, marker='s')
axs[1, 1].scatter(e[:admitted*2+1], f[:admitted*2+1], c=c2, marker='s')
axs[1, 1].set_title('Female/Square')
```

```python
# Labels and Axes

##for ax in axs.flat:
##    ax.set(xlabel='null', ylabel='y-label')

# Hide x labels and tick labels for top plots and y ticks for right plots.
for ax in axs.flat:
    ax.label_outer()
```
